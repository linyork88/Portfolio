```
DC=function(mt,list1){ #直接選取
  mt1=mt
  mt1[-list1]=0
  mt1[list1]=1
  return(mt1)
}
```

```
InDC=function(mt,S,S1,list1){ #剩餘選取
  mt2=mt
  mt2[-list1]=0
  mt2[list1]=1
  list2=c(1:36)
  list3=list2[-list1]
  #print(c("從這些牌",list(list3),"選",S1,"張"),sep='',end='')
  if (S1==1 & S+S1==36){
    mt2[list3]=1
    list4=list3
  }
  else{
    list4=sort(sample(list3,S1,replace = FALSE))
  }
  #print(c("選擇",list4))
  mt2[list4]=1
  return(mt2)
}
```

```
Choose=function(S,S1){ #在S張牌且聽牌後摸S1張的設置下
  n=0 #直接連線數
  n1=0 #聽牌數
  N=0 #聽牌後連線數
  mt=matrix(1:36,6)
  list1=sort(sample(1:36,S,replace=FALSE))
  DirC=DC(mt,list1) #直接選取
  #print(DirC)
    for(i in c(1:6)){ #直接連線情形
      if(sum(DirC[i,])==6){
        n=n+1
      }
      if(sum(DirC[,i])==6){
        n=n+1
      }
    }
    if(sum(DirC[1,1],DirC[2,2],DirC[3,3],DirC[4,4],DirC[5,5],DirC[6,6])==6){
      n=n+1
    }
    if(sum(DirC[6,1],DirC[5,2],DirC[4,3],DirC[3,4],DirC[2,5],DirC[1,6])==6){
      n=n+1
    }
    if(n==0){ #聽牌情形
      for(i in c(1:6)){
        if(sum(DirC[i,])==5){
        n1=n1+1
        }
        if(sum(DirC[,i])==5){
          n1=n1+1
        }
      }
      if(sum(DirC[1,1],DirC[2,2],DirC[3,3],DirC[4,4],DirC[5,5],DirC[6,6])==5){
      n1=n1+1
      }
      if(sum(DirC[6,1],DirC[5,2],DirC[4,3],DirC[3,4],DirC[2,5],DirC[1,6])==5){
      n1=n1+1
      } 
    }
  if(n1>=1){ #聽牌後摸牌情形
  InDirC=InDC(mt,S,S1,list1) #剩餘選取
  #print(InDirC)
    for(i in c(1:6)){
    if(sum(InDirC[i,])==6){
      N=N+1
    }
    if(sum(InDirC[,i])==6){
      N=N+1
    }
    }
    if(sum(InDirC[1,1],InDirC[2,2],InDirC[3,3],InDirC[4,4],InDirC[5,5],InDirC[6,6])==6){
      N=N+1
    }
    if(sum(InDirC[6,1],InDirC[5,2],InDirC[4,3],InDirC[3,4],InDirC[2,5],InDirC[1,6])==6){
      N=N+1
    }
  }
  return(c(n,n1,N))
}
```

```
Test=function(S,S1,G){
X=0
Y=0
Z=0
for(i in 1:G){
  temp=c(Choose(S,S1))
  if (temp[1]>0){
    X=X+1
  }
  if (temp[2]>0){
    Y=Y+1
  }
  if (temp[3]>0){
    Z=Z+1
  }
}
print(paste("在抽取",S,"張","且聽牌後抽取",S1,"張的設置下執行",i,"次",sep = ""))
print(paste("直接連線局數=",X,"   ","直接連線率為",(round(X/G,4)*100),"%",sep = ""))
print(paste("聽牌局數(排除直連)=",Y,"   ","聽牌率(排除直連)為",(round(Y/G,4)*100),"%",sep = ""))
print(paste("聽牌後連線局數=",Z,"   ","聽牌後連線率為",round(Z/Y*100,4),"%","   ","聽連佔整體為",round(Z/G*100,4),"%",sep = ""))
print(paste("開局結束局數=",G-(X+Y),"   ","開局結束率為",(round((G-(X+Y))/G,4)*100),"%",sep = "")) #意指連聽牌都沒有的局數
print(paste("整體貢龜局數=",G-(X+Z),"   ","整體貢龜率為",(round((G-(X+Z))/G,4)*100),"%",sep = "")) #意指任何形式連線都沒有的局數
}
```
