```python
#@title
import random
import math
import numpy as np
import pandas as pd
```

```python
#@title
# C N 取 fq
def Choose(N,fq):
    a=0
    a=(math.factorial(N)/math.factorial(N-fq))/math.factorial(fq)
    return(a)
```

```python
#@title
def billseperate(billlist,budget,P=100):
    n=0
    N=len(billlist)
    fq=math.ceil(budget/(np.mean(billlist))) #適合帳目比數
    total=sum(billlist) #資料數總和
    bill=[]
    billtotal=0
    while total> billtotal:
        temp= sorted(random.sample(billlist,fq))
        if temp in bill:
            continue
        else:
            bill.append(temp)
            n+=1
            #print("新增組合:",temp)
            bill.sort()
            #print(bill)
        if n==Choose(N,fq) or n==int((Choose(N,fq)*P)/100):
            break
    print("發揮",P,"%的效能","從",N,"筆帳目中取",fq,"筆的排列組合=",n)

    x=0
    max_bill=[]
    s=0
    if n==int((Choose(N,fq)*P)/100):
      s=n
    else:
      s=Choose(N,fq)
    for i in range(0,int(s)):
        if sum(bill[i])<=budget:
            x+=1
            max_bill.append((bill[i]))
            #print("可拆帳金額第%d筆="%(x),bill[i],"=",sum(bill[i]))
    for k in range(0,len(max_bill)):
        if sum(sorted(max_bill, reverse=True)[0])==sum(sorted(max_bill, reverse=True)[k]):
            print("最佳拆法",sorted(max_bill, reverse=True)[k],"=",sum(sorted(max_bill, reverse=True)[k]))
```
